## <b>KGLiDS APIs</b>

```python
import pandas as pd
```

```python
from gnn_applications.OnDemandDataPrep.api import KGLiDSDataPrep
kglids_dataprep = KGLiDSDataPrep(endpoint='http://localhost:7200', db='kaggle_demo')
```

```python
df = pd.read_csv('experiments/OnDemandDataPrep_experiments/Cleaning_datasets/hepatitis.csv')
```

```python
clean_df = kglids_dataprep.apply_cleaning_operations(cl_op,df)
```
